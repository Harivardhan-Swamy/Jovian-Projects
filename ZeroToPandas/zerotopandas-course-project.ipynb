## Exploratory Analysis and Visualization

**TODO** - Visual representation of data is the easiest to understand and interpret.
So appropriate representation is necessary.This is done using sns and matplot to display the graphs.



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (15,10)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

**TODO** - Explore one or more columns by plotting a graph below, and add some explanation about it

````python
plt.bar(vid_df.Year,vid_df.Global_Sales)
plt.xlabel='Years'
plt.ylabel='Global Sales'
````

Plots a graph of total sales versus the year in which the game was sold

**TODO** - Explore one or more columns by plotting a graph below, and add some explanation about it

````python
plt.plot(vid_df['Genre'].value_counts()[:300])
plt.xticks(rotation="vertical")
````

Plots the number of games in each genre versus  the genre

**TODO** - Explore one or more columns by plotting a graph below, and add some explanation about it

````python
import numpy as np
genre=vid_df.Genre.unique()
genre.sort()
rank=vid_df.groupby('Genre').count()['Rank']
rank=np.array(rank)
print(rank)
sns.barplot(rank,genre,hue=genre)
````

Plots the number of games in each genre versus  the genre

**TODO** - Explore one or more columns by plotting a graph below, and add some explanation about it

````python
plat=vid_df.Platform.unique()
plat.sort()
rank=vid_df.groupby('Platform').count()['Rank']
rank=np.array(rank)
plt.plot(plat,rank)
````

The number of games released for each console

**TODO** - Explore one or more columns by plotting a graph below, and add some explanation about it

Number of games published by each publisher

````python
ar = np.array(vid_df.groupby('Publisher')['Rank'].count())
ar.sort()
ar=ar[::-1]
ar=ar[:20]
pbs=vid_df.groupby("Publisher").count()
pbs=pbs.sort_values(by='Name',ascending=False)['Name'].head(20)
pbs.index
sns.barplot(pbs.index,ar)
plt.plot(pbs.index,ar, 'b-o', linewidth=4, markersize=12, markeredgewidth=4, markeredgecolor='navy')
plt.xticks(rotation="vertical")
````
